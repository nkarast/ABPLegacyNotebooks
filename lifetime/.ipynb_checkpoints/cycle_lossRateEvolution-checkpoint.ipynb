```python
# Calculate Loss Rate
dt_b1_h = (pd.Timestamp(FBCT_BEAM1_DF[filled_slots_b1].index.values[1]) - pd.Timestamp(FBCT_BEAM1_DF[filled_slots_b1].index.values[0])).seconds/3600.
dt_b2_h = (pd.Timestamp(FBCT_BEAM2_DF[filled_slots_b2].index.values[1]) - pd.Timestamp(FBCT_BEAM2_DF[filled_slots_b2].index.values[0])).seconds/3600.

FBCT_B1_NORM_LOSS_RATE = -1.0*((FBCT_BEAM1_DF[filled_slots_b1].diff(axis=0)/dt_b1_h)*(1./FBCT_BEAM1_DF[filled_slots_b1]))[1:]
FBCT_B2_NORM_LOSS_RATE = -1.0*((FBCT_BEAM2_DF[filled_slots_b2].diff(axis=0)/dt_b2_h)*(1./FBCT_BEAM2_DF[filled_slots_b2]))[1:]
```

```python
dt_b1_ft_h = (pd.Timestamp(FBCT_BEAM1_FT_DF.index.values[1]) - pd.Timestamp(FBCT_BEAM1_FT_DF.index.values[0])).seconds/3600.
dt_b2_ft_h = (pd.Timestamp(FBCT_BEAM2_FT_DF.index.values[1]) - pd.Timestamp(FBCT_BEAM2_FT_DF.index.values[0])).seconds/3600.

FBCT_B1_FT_NORM_LOSS_RATE = -1.0*((FBCT_BEAM1_FT_DF.diff(axis=0)/dt_b1_h)*(1./FBCT_BEAM1_FT_DF))[1:]
FBCT_B2_FT_NORM_LOSS_RATE = -1.0*((FBCT_BEAM2_FT_DF.diff(axis=0)/dt_b2_h)*(1./FBCT_BEAM2_FT_DF))[1:]
```

```python
empty_slots_b1 = np.where(b1_fill_pattern != 1)[0]
empty_slots_b2 = np.where(b2_fill_pattern != 1)[0]
```

```python
FBCT_B1_FT_NORM_LOSS_RATE[empty_slots_b1] = 0
FBCT_B2_FT_NORM_LOSS_RATE[empty_slots_b2] = 0
```
