# Outlier Detection for (BCMS) Emittances,  
~ N. Karastathis


The aim of this notebook is to 
- give an example on **basic outlier cleaning** in the existing the statistical population with **CL-based methods**;
- automatic **outlier detection** based on **supervised classification model**;

The goal of the study is to build a robust model that identifies drops on the efficiency of the emittance measurements, which being an optical system can drift with time. Fast identification of faults can lead to fast reaction and therefore increase the usability and dependability of the system.

The dataset is generated by the 2017/2018 LHC fills and it takes into account the BSRT emittance measurements and the fBCT intensity measurements. To randomize the selection a shuffle and random sampling has been applied to the fill list. The initial data are taken by the **serialized** files generated by the [LuminosityFollowUp](https://github.com/nkarast/LuminosityFollowUp) which transforms the **RDB** stored Accelerator Data to flat observables. 

The contents are

1. Randomization and **data selection**
2. **Basic Data Preparation (cleaning, missing data, re-formatting, flattening the dataset)**
3. **Feature engineering** to combine the available information and reduce the **dimensionality**
4. An **exploratory data analysis (EDA)** is performed on the dataset to identify possible **correlations** of the features to minimize the dimensionality of input data by capitalizing on the **information** provided.
5. The **machine learning part** checks for the **accuracy** and **AUC ROC** of various **supervised classification** algorithms. A **cross-validation** is performed in a grid in order to optimise the model **hyper-parameters**
6. In order to **report the results** the information is passed in **GitHub** where a **Pages** is created with **Jekyll**.

```python
# Create a function to clean a DF based on the IQR
def filter_on_iqr(df, column, whisker_width=1.5):
    # Calculate Q1, Q2 and IQR
    q1 = df[column].quantile(0.25)                 
    q3 = df[column].quantile(0.75)
    iqr = q3 - q1
    # Apply filter with respect to IQR, including optional whiskers
    filter = (df[column] >= q1 - whisker_width*iqr) & (df[column] <= q3 + whisker_width*iqr)
    return df.loc[filter], df.loc[~filter]
```

```python
# calculate the IRQ for each column
whisker_width = 1.5

q1_1  = df['injEmitH_bunch_b1'].quantile(0.25); q3_1 = df['injEmitH_bunch_b1'].quantile(0.75); iqr_1 = q3_1 - q1_1;
q1_2  = df['injEmitH_bunch_b2'].quantile(0.25); q3_2 = df['injEmitH_bunch_b2'].quantile(0.75); iqr_2 = q3_2 - q1_2;
q1_3  = df['injEmitV_bunch_b1'].quantile(0.25); q3_3 = df['injEmitV_bunch_b1'].quantile(0.75); iqr_3 = q3_3 - q1_3;
q1_4  = df['injEmitV_bunch_b2'].quantile(0.25); q3_4 = df['injEmitV_bunch_b2'].quantile(0.75); iqr_4 = q3_4 - q1_4;

q1_5  = df['sbEmitH_bunch_b1'].quantile(0.25); q3_5 = df['sbEmitH_bunch_b1'].quantile(0.75); iqr_5 = q3_5 - q1_5;
q1_6  = df['sbEmitH_bunch_b2'].quantile(0.25); q3_6 = df['sbEmitH_bunch_b2'].quantile(0.75); iqr_6 = q3_6 - q1_6;
q1_7  = df['sbEmitV_bunch_b1'].quantile(0.25); q3_7 = df['sbEmitV_bunch_b1'].quantile(0.75); iqr_7 = q3_7 - q1_7;
q1_8  = df['sbEmitV_bunch_b2'].quantile(0.25); q3_8 = df['sbEmitV_bunch_b2'].quantile(0.75); iqr_8 = q3_8 - q1_8;

mfilter  = (df['injEmitH_bunch_b1'] >= q1_1 - whisker_width*iqr_1) & (df['injEmitH_bunch_b1'] <= q3_1 + whisker_width*iqr_1) & (df['injEmitH_bunch_b2'] >= q1_2 - whisker_width*iqr_2) & (df['injEmitH_bunch_b2'] <= q3_2 + whisker_width*iqr_2) & (df['injEmitV_bunch_b1'] >= q1_3 - whisker_width*iqr_3) & (df['injEmitV_bunch_b1'] <= q3_3 + whisker_width*iqr_3) & (df['injEmitV_bunch_b2'] >= q1_4 - whisker_width*iqr_4) & (df['injEmitV_bunch_b2'] <= q3_4 + whisker_width*iqr_4) & (df['sbEmitH_bunch_b1']  >= q1_5 - whisker_width*iqr_5) & (df['sbEmitH_bunch_b1']  <= q3_5 + whisker_width*iqr_5) & (df['sbEmitH_bunch_b2']  >= q1_6 - whisker_width*iqr_6) & (df['sbEmitH_bunch_b2']  <= q3_6 + whisker_width*iqr_6) & (df['sbEmitV_bunch_b1']  >= q1_7 - whisker_width*iqr_7) & (df['sbEmitV_bunch_b1']  <= q3_7 + whisker_width*iqr_7) & (df['sbEmitV_bunch_b2']  >= q1_8 - whisker_width*iqr_8) & (df['sbEmitV_bunch_b2']  <= q3_8 + whisker_width*iqr_8)
df_clean = df.loc[mfilter]
df_dirty = df.loc[~mfilter]
```

Setup a classification algorithm to identify clean or not events

## 3.1 Split Train-Test

```python
from sklearn.model_selection import train_test_split
```

```python
dfX = dfbcms[['injEmitH_bunch_b1', 'injEmitH_bunch_b2', 'injEmitV_bunch_b1', 'injEmitV_bunch_b2',  'sbEmitH_bunch_b1', 'sbEmitH_bunch_b2', 'sbEmitV_bunch_b1', 'sbEmitV_bunch_b2', 'injIntensity_bunch_b1', 'injIntensity_bunch_b2',]]
dfy = dfbcms['clean']
```

```python
X_train, X_test, y_train, y_test = train_test_split(dfX, dfy, test_size=0.33, random_state=42)
```

## 3.2 Model Selection with Cross-Validation (CV)

```python
from sklearn.linear_model import LogisticRegression  # Logistic Regression
from sklearn.neighbors import KNeighborsClassifier # kn classifier
from sklearn.tree import DecisionTreeClassifier # Decision Tree
from sklearn.svm import SVC # support vector classifier
from sklearn.ensemble import AdaBoostClassifier # create Ada boosted classifier
#  CV and scoring
from sklearn.model_selection import cross_val_score
from sklearn import metrics
# time
import time
```

A first step is to test some basic models

```python
models = []
```

```python
logreg = LogisticRegression()
models.append(logreg)
```

```python
kn = KNeighborsClassifier()
models.append(kn)
```

```python
dt = DecisionTreeClassifier(criterion='gini', splitter='best', max_depth=5, min_samples_leaf=1000)
models.append(dt)
```

```python
svc = SVC()
models.append(svc)
```

```python
bdt = AdaBoostClassifier(dt)
models.append(bdt)
```

```python
model_names = ['Logistic Regression', 'K-Nearest Neighbours', 'Decision Tree', 'Support Vector', 'Boosted Decision Tree']
```

```python
for idx, m_model in enumerate(models):
    start      = time.time()
    tmp_scores = cross_val_score(m_model, X_test, y_test, cv=5, scoring='accuracy')
    end        = time.time()
    model_name = model_names[idx]
    print("-----------------------------------------------")
    print("Model\t:\t{}".format(model_name))
    print("Time\t:\t{:.5f}".format(end-start))
    print("Accuracy\t:\t{:.3f} +\- {:.3f}".format(tmp_scores.mean(), tmp_scores.std()))
```

## 3.2 GridSearchCV 
Let's select the BDT and do a GridSearchCV with 5-fold CV for hyper-parameter optimization

```python
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import classification_report
```

```python
A   = DecisionTreeClassifier(criterion='gini', splitter='best', max_depth=5, min_samples_leaf=1000, random_state=42)
BDT = AdaBoostClassifier(base_estimator=A)
```

```python
grid = {"base_estimator__criterion" : ['gini', 'entropy'],
        "base_estimator__max_depth" : [ 5, 10, 15, 20],
        "base_estimator__min_samples_leaf" : [500, 1000, 1500, 2000],
        "n_estimators": [10, 50, 100, 150, 200]
}

scores = ['accuracy', 'roc_auc']
for score in scores:
    print("# Optimizing Hyper-parameters for {}".format(score))
    clf = GridSearchCV(BDT, grid, cv=5, scoring='{}'.format(score), verbose=3, n_jobs=4 )
    clf.fit(X_train, y_train)

    print("Best parameters found:")
    print("")
    print(clf.best_params_)
    print("")
    print("Grid scores:")
    print("")
    for mean, std, params in zip(clf.cv_results_['mean_test_score'], clf.cv_results_['std_test_score'], clf.cv_results_['params']):
        print("{:0.3f} (+/-{:0.03f}) for {}".format(mean, std, params))
    print("")

    print("Classification report:")
    print("")
    print("N.B. Training and test samples are used.")
    print("")
    y_true, y_pred = y_test, clf.predict(X_test)
    print(classification_report(y_true, y_pred))
    print("------------------------------------------------------")
```

The best configuration for accuracy is : 
    {'n_estimators': 200, 'base_estimator__criterion': 'gini', 'base_estimator__max_depth': 5, 'base_estimator__min_samples_leaf': 1500}

The best configuration for roc_auc is : 
    {'n_estimators': 10, 'base_estimator__criterion': 'gini', 'base_estimator__max_depth': 5, 'base_estimator__min_samples_leaf': 1500}

Therefore the difference comes from the Boosting Hyper parameters. Testing a DT with the above configuration.
