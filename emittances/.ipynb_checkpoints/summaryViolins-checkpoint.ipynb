# Summary Violin Plots 
~ N. Karastathis (2017)

This is a summary visualisation notebook. After having created the dataframes for the emittance data (`createDataFrames.ipynb`) and possibly cleaned them (`cleanDataFrames.ipynb`) you can visualize the important information using this notebook. 

The code visualizes both beams and both planes making use of the hue-ed violinplots from Seaborn. You can have the emittances at different LHC cycle instances, the relative change and the growth, as well as information about the bunch-by-bunch intensity and brightness evolution.


Have fun!


```python
import numpy as np
import glob
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import pickle, gzip
```

```python
FILEDIR = '/eos/user/n/nkarast/LHC/2018/FollowUp/pickles/'
summary = FILEDIR+'summaryEmittanceDataFrame_2018_periodB.pkl.gz' #FILEDIR+'/2017/summaryEmittanceDataFrame.h5'# 
growth  = FILEDIR+'summaryEmittanceGrowthDataFrame_2018_periodB.pkl.gz'
EOSDIR  = '/eos/project/l/lhc-lumimod/LuminosityFollowUp/2018/procdata/'
```

### Emittances at Start of Stable Beams

```python
# ########  LOAD THE PICKLE FILE
# ##df = pd.read_hdf(summary, 'df')
# with gzip.open(summary, 'rb') as fid:
#     df = pickle.load(fid)

    

#########  BASIC SELECTION
df2 = df_sum[(df_sum['Cycle']=='Start Stable') & (~df_sum['Fill'].isin(blacklist)) & (df_sum['Kind']=='Measurement')]

with sns.axes_style("whitegrid"):
    #########  SETUP THE PLOT
    fig_emit = plt.figure(1, figsize=(20,9))
    ax_b1 = plt.subplot(211)
    ax_b2 = plt.subplot(212, sharex=ax_b1, sharey=ax_b1)
    sns.set_style("whitegrid")
    ax_b1 = sns.violinplot(x="Fill", y="Emittance", hue="Plane", data=df2[df2['Beam']=='B1'],  palette=['#71aaf2', '#ff8a3d'], lw=0,
                           split=True, order=flist, ax=ax_b1, cut=0)
    ax_b2 = sns.violinplot(x="Fill", y="Emittance", hue="Plane", data=df2[df2['Beam']=='B2'],  palette=['#71aaf2', '#ff8a3d'], lw=0,
                           split=True, order=flist, ax=ax_b2, cut=0)
    # Legend handling
    ax_b1.legend_.set_title('')
    ax_b1.legend_.set_frame_on(True)
    ax_b1.legend_.set_alpha(0.8)
    ax_b2.legend_.set_visible(False)
    
    # Adding total intensity
#     ax_b1_bunch = ax_b1.twiny()
#     ax_b1_bunch.set_xlim(ax_b1.get_xlim())
#     ax_b1_bunch.set_xticks(ax_b1.get_xticks())
#     x_bunches = ["{}b".format(df2['Bunches'][df2['Fill']==ifil].unique()[0]) for ifil in flist]
#     #ax_b1_bunch.xaxis.set_major_locator(plt.MaxNLocator(len(x_bunches)))
#     ax_b1_bunch.set_xticklabels(x_bunches)

    ######### GET VALUES FOR LABELS
    mean_values_b1 = []
    std_values_b1  = []
    for nbunches in flist:
         for nplane in ['Horizontal', 'Vertical']:
                mean_values_b1.append(df2['Emittance'][(df2['Fill']==nbunches) & (df2['Plane']==nplane) & (df2['Beam']=='B1')].mean())
                std_values_b1.append(df2['Emittance'][(df2['Fill']==nbunches) & (df2['Plane']==nplane) & (df2['Beam']=='B1')].std())

    mean_values_b2 = []
    std_values_b2 = []
    for nbunches in flist:
         for nplane in ['Horizontal', 'Vertical']:
                mean_values_b2.append(df2['Emittance'][(df2['Fill']==nbunches) & (df2['Plane']==nplane) & (df2['Beam']=='B2')].mean())
                std_values_b2.append(df2['Emittance'][(df2['Fill']==nbunches) & (df2['Plane']==nplane) & (df2['Beam']=='B2')].std())

    mean_labels_b1 = [str(np.round(s, 2)) for s in mean_values_b1]
    mean_labels_b2 = [str(np.round(s, 2)) for s in mean_values_b2]
    pos = range(len(mean_labels_b1))

    ######### SET AXIS LABELS
    ax_b1.set_ylabel('B1 $\mathbf{\epsilon_{n}}$ [$\mathbf{\mu}$m]', fontsize=18, fontweight='bold')
    ax_b2.set_ylabel('B2 $\mathbf{\epsilon_{n}}$ [$\mathbf{\mu}$m]', fontsize=18, fontweight='bold')
    ax_b2.set_xlabel('Fill Number', fontsize=16, fontweight='bold', labelpad=20)
    ax_b1.set_xlabel('')
    ax_b1.set_title('Emittance at the Start of Stable Beams', fontsize=20, fontweight='bold', y=1.10)

    ######### SET THE MEAN VALUES TAGS
    # for tick,label in zip(pos,ax_b1.get_xticklabels()):
    # 	ax_b1.text(pos[tick]-0.3, mean_values_b1[2*tick],   mean_labels_b1[2*tick],    horizontalalignment='center', size='small', color='k')
    # 	ax_b1.text(pos[tick]+0.3, mean_values_b1[2*tick+1], mean_labels_b1[2*tick+1],  horizontalalignment='center', size='small', color='k')
    # for tick,label in zip(pos,ax_b2.get_xticklabels()):
    # 	ax_b2.text(pos[tick]-0.3, mean_values_b2[2*tick],   mean_labels_b2[2*tick],    horizontalalignment='center', size='small', color='k')
    # 	ax_b2.text(pos[tick]+0.3, mean_values_b2[2*tick+1], mean_labels_b2[2*tick+1],  horizontalalignment='center', size='small', color='k')

    ######### ROTATE X-TICKLABELS
    for tick in ax_b1.get_xticklabels():
        tick.set_rotation(90)
    for tick in ax_b2.get_xticklabels():
        tick.set_rotation(90)

    ######### SET AXIS LIMITS
    ax_b1.set_ylim(0.5,4)
    ax_b2.set_ylim(0.5,4)

    ax_b1.grid('on')
    ax_b2.grid('on')


    plt.setp(ax_b1.get_xticklabels(), fontsize=15)
    plt.setp(ax_b1.get_yticklabels(), fontsize=15)
    plt.setp(ax_b2.get_xticklabels(), fontsize=15)
    plt.setp(ax_b2.get_yticklabels(), fontsize=15)


    plt.tight_layout()

# SAVE??
# fig_emit.savefig('plots_nikos/emit_atStable_{}.png'.format(datetime.datetime.now().strftime("%Y%m%d")), dpi=300)

##### GET SOME STATS    
horiz1 = []
vert1  = []
horiz2 = []
vert2  = []

for indx in range(len(mean_values_b1)):
    if indx%2 == 0:
        horiz1.append(mean_values_b1[indx])
        horiz2.append(mean_values_b2[indx])
    else:
        vert1.append(mean_values_b1[indx])
        vert2.append(mean_values_b2[indx])
        
        
print 'MEAN VALUES +/- RMS OF ALL FILLS:'
print 'B1H :', np.around(np.nanmean(np.array(horiz1)),3)  , '+/-', np.around(np.nanstd(np.array(horiz1)),3 )
print 'B1V :', np.around(np.nanmean(np.array(vert1)),3)   , '+/-', np.around(np.nanstd(np.array(vert1)),3 )
print 'B2H :', np.around(np.nanmean(np.array(horiz2)),3)  , '+/-', np.around(np.nanstd(np.array(horiz2)),3 )
print 'B2V :', np.around(np.nanmean(np.array(vert2)),3)   , '+/-', np.around(np.nanstd(np.array(vert2)),3 )
```
